kmeans = pd.read_csv("creditcard.csv")

```python
kmeans.info()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kmeans_features = kmeans.drop('class',axis = 1)
kmeans_target = kmeans['class']
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(kmeans_features, kmeans_target, test_size=0.30, random_state=42)
```

```python
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train2,y_train2)
    pred_i = knn.predict(X_test2)
    error_rate.append(np.mean(pred_i != y_test2))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knneig = KNeighborsClassifier(n_neighbors=2)
knneig.fit(X_train2, y_train2)
pred_knneigh = knneig.predict(X_test2)
score_knneigh_before = accuracy_score(y_test2, pred_knneigh)
print("Score KNeighnors :",score_knneigh_before)
print(classification_report(y_test2, pred_knneigh))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100,random_state = 42)
rfc.fit(X_train2, y_train2)
rfc_pred = rfc.predict(X_test2)
print(accuracy_score(y_test2,rfc_pred))
print(classification_report(y_test2,rfc_pred))
```

```python
from sklearn.svm import SVC
```

```python
#param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 
#from sklearn.model_selection import GridSearchCV
grid = SVC(probability = True, random_state = 42)
# May take awhile!
grid.fit(X_train2,y_train2)
grid_predictions = grid.predict(X_test2)
print(accuracy_score(y_test2,grid_predictions))
print(classification_report(y_test2,grid_predictions))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gauss_clf = GaussianNB()
gauss_clf.fit(X_train2, y_train2)
pred_gauss = gauss_clf.predict(X_test2)
print(accuracy_score(y_test2, pred_gauss))
print(classification_report(y_test2,pred_gauss))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train2,y_train2)
predictions = logmodel.predict(X_test2)
print(accuracy_score(y_test2, predictions))
print(classification_report(y_test2,predictions))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
etc = ExtraTreesClassifier(n_estimators=250, random_state=42)
etc.fit(X_train2,y_train2)
pred_etc = etc.predict(X_test2)
print(accuracy_score(y_test2, pred_etc))
print(classification_report(y_test2,pred_etc))
```

```python
gbc = GradientBoostingClassifier(loss='exponential', learning_rate=0.03, n_estimators=75 , max_depth=6)
gbc.fit(X_train2,y_train2)
pred_gbc = gbc.predict(X_test2)
print(accuracy_score(y_test2, pred_gbc))
print(classification_report(y_test2,pred_gbc))
```

```python
ada = AdaBoostClassifier(n_estimators=100, random_state=42)
ada.fit(X_train2,y_train2)
pred_ada = ada.predict(X_test2)
print(accuracy_score(y_test2, pred_ada))
print(classification_report(y_test2,pred_ada))
```

```python
import xgboost as xgb
xg = xgb.XGBClassifier(objective="binary:logistic", random_state=42, eval_metric="auc")
xg.fit(X_train2,y_train2)
pred_xg = xg.predict(X_test2)
print(accuracy_score(y_test2, pred_xg))
print(classification_report(y_test2,pred_xg))
```

```python
from sklearn import tree
```

```python
dect = tree.DecisionTreeClassifier(random_state= 42)
dect.fit(X_train2,y_train2)
pred_dect = dect.predict(X_test2)
print(accuracy_score(y_test2, pred_dect))
print(classification_report(y_test2,pred_dect))
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
```

```python
r_probs = [0 for _ in range(len(y_test2))]
KNN_probs = knneig.predict_proba(X_test2)
RF_probs = rfc.predict_proba(X_test2)
SVM_probs = grid.predict_proba(X_test2)
NB_probs = gauss_clf.predict_proba(X_test2)
LR_probs = logmodel.predict_proba(X_test2)
ETC_probs = etc.predict_proba(X_test2)
GBC_probs = gbc.predict_proba(X_test2)
ADA_probs = ada.predict_proba(X_test2)
XG_probs = xg.predict_proba(X_test2)
DT_probs = dect.predict_proba(X_test2)

KNN_probs = KNN_probs[:, 1]
RF_probs = RF_probs[:, 1]
SVM_probs = SVM_probs[:, 1]
LR_probs = LR_probs[:, 1]
NB_probs = NB_probs[:, 1]
ETC_probs = ETC_probs[:,1]
GBC_probs = GBC_probs[:,1]
ADA_probs = ADA_probs[:,1]
XG_probs = XG_probs[:,1]
DT_probs = DT_probs[:,1]

r_auc = roc_auc_score(y_test2, r_probs)
KNN_auc = roc_auc_score(y_test2, KNN_probs)
RF_auc = roc_auc_score(y_test2, RF_probs)
SVM_auc = roc_auc_score(y_test2, SVM_probs)
LR_auc = roc_auc_score(y_test2, LR_probs)
NB_auc = roc_auc_score(y_test2, NB_probs)
ETC_auc = roc_auc_score(y_test2, ETC_probs)
GBC_auc = roc_auc_score(y_test2, GBC_probs)
ADA_auc = roc_auc_score(y_test2, ADA_probs)
XG_auc = roc_auc_score(y_test2, XG_probs)
DT_auc = roc_auc_score(y_test2, DT_probs)

r_fpr, r_tpr, _ = roc_curve(y_test2, r_probs)
KNN_fpr, KNN_tpr, _ = roc_curve(y_test2, KNN_probs)
RF_fpr, RF_tpr, _ = roc_curve(y_test2, RF_probs)
SVM_fpr, SVM_tpr, _ = roc_curve(y_test2, SVM_probs)
LR_fpr, LR_tpr, _ = roc_curve(y_test2, LR_probs)
NB_fpr, NB_tpr, _ = roc_curve(y_test2, NB_probs)
ETC_fpr, ETC_tpr, _ = roc_curve(y_test2, ETC_probs)
GBC_fpr, GBC_tpr, _ = roc_curve(y_test2, GBC_probs)
ADA_fpr, ADA_tpr, _ = roc_curve(y_test2, ADA_probs)
XG_fpr, XG_tpr, _ = roc_curve(y_test2, XG_probs)
DT_fpr, DT_tpr, _ = roc_curve(y_test2, DT_probs)
```

```python
plt.figure(figsize=(8,6))
plt.plot(r_fpr, r_tpr, linestyle='--')
#plt.plot(rf_fpr, rf_tpr, marker='.', label='Random Forest (AUROC = %0.3f)' % rf_auc)
plt.plot(KNN_fpr, KNN_tpr, label='KNN (AUROC = %0.3f)' % KNN_auc)
plt.plot(RF_fpr, RF_tpr, label='RF (AUROC = %0.3f)' % RF_auc)
plt.plot(SVM_fpr, SVM_tpr, label='SVM (AUROC = %0.3f)' % SVM_auc)
plt.plot(LR_fpr, LR_tpr, label='LR (AUROC = %0.3f)' % LR_auc)
plt.plot(NB_fpr, NB_tpr, label='NB (AUROC = %0.3f)' % NB_auc)

# Title
plt.title('ROC Plot')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# Show legend
plt.legend() # 
# Show plot
plt.show()
```
